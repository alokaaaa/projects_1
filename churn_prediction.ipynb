```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
db = pd.read_csv('Churn_Modelling.csv')
```

```python
db.head()
```

```python
db=db.drop(columns=["RowNumber","CustomerId","Surname"])
```

```python
columns_to_be_encoded=["Geography","Gender"]
```

```python
from sklearn.preprocessing import OneHotEncoder
X = db[columns_to_be_encoded]
encoder = OneHotEncoder()
X_encoded = encoder.fit_transform(X)
X_encoded = pd.DataFrame(X_encoded.toarray(), columns=encoder.get_feature_names_out(columns_to_be_encoded))
db_with_encoded = pd.concat([db, X_encoded], axis=1)
```

```python
db_with_encoded=db_with_encoded.drop(columns=["Geography","Gender"])
```

```python
db_with_encoded.head()
```

```python
X=db_with_encoded.drop(columns=["Exited"])
y=db_with_encoded["Exited"]
```

```python
from sklearn.linear_model import LogisticRegression
my_model=LogisticRegression(random_state=42)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
my_model.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
y_predict=my_model.predict(X_test)
accuracy = accuracy_score(y_test, y_predict)
print(f"Accuracy: {accuracy*100}%")
```
